```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/sales_data_with_discounts.csv")
df.head()
df.dtypes
num=df.select_dtypes(include=['int64','float64'])
stats={'Mean': num.mean(),
       'Median': num.median(),
       'Mode': num.mode().iloc[0],
       'Max': num.max(),
       'Min': num.min(),
       'Standard Deviation': num.std(),
       'Variance': num.var()}
stats_df=pd.DataFrame(stats)
stats_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
num=df.select_dtypes(include=['int64','float64'])
for col in num.columns:
  if df.dtypes[col] in ['int64','float64']:
    plt.figure(figsize=(15,5))
    sn.histplot(df[col],color='blue', edgecolor='black',bins=30)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
    kurtvalue=df[col].kurt()
    print("kurtosis of",col ,":" ,kurtvalue)
    skewvalue=df[col].skew()
    print("skewness of",col ,":" ,skewvalue)
```

```python
import warnings
warnings.filterwarnings('ignore')
for col in num.columns:
  if df.dtypes[col] in ['int64','float64']:
    plt.figure(figsize=(15,5),edgecolor="pink")
    plt.boxplot(df[col],vert=False)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
Q1 = np.quantile(df[col],0.25)
Q3 = np.quantile(df[col],0.75)
IQR = Q3 - Q1
lowerOL = Q1 - 1.5 * IQR
upperOL = Q3 + 1.5 * IQR
extrupper = Q1 - 3 * IQR
extrlower = Q3 + 1.5 * IQR
outliers = num[(num < lowerOL) | (num> upperOL)]
extreme_values = num[(num < extrlower) | (num > extrupper)]
outliers.isnull().sum()
extreme_values.isnull().sum()
if not outliers.empty:
  print(f"Outliers for {col}:")
  print(outliers)
if not extreme_values.empty:
  print(f"Extreme values for {col}:")
  print(extreme_values)
```

```python
numc = df.select_dtypes(include=['object'])
for col in numc.columns:
  if df.dtypes[col] in ['object']:
    plt.figure(figsize=(5,5))
    df[col].value_counts().plot(kind='bar',color="pink",edgecolor="purple")
    plt.title(f"Bar plot of {col}")
    plt.xlabel(col)
    plt.ylabel("count")
```

```python
mean = num.mean()
sd = num.std()
standardized_num = (num - mean) / sd
print("Before Standardization:")
print(num.describe())
print("\nAfter Standardization:")
print(standardized_num.describe())
for col in num.columns:
  plt.figure(figsize=(15, 5))
  plt.subplot(1, 2, 1)
  sn.histplot(df[col], color='blue', edgecolor='black', bins=10)
  plt.title(f"Histogram of {col} (Before Standardization)")
  plt.xlabel(col)
  plt.ylabel("Frequency")

  plt.subplot(1, 2, 2)
  sn.histplot(standardized_num[col], color='red', edgecolor='black', bins=30)
  plt.title(f"Histogram of {col} (After Standardization)")
  plt.xlabel(col)
  plt.ylabel("Frequency")
  plt.show()
```
